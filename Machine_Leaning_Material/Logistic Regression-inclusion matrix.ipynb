```python
#applicable in the case where we want ouput in the form of 0 and 1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#splitting data set
from sklearn.model_selection import train_test_split
#for feature scaling
from sklearn.preprocessing import StandardScaler
#for modeling classifier
from sklearn.linear_model import LogisticRegression
#lib for confusion 
#sklearn main library, sub libray metrics and clas confusion_matrix
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv(r'/Users/divyashahi/OneDrive - Adobe/Machine_Leaning_Material/Ads.csv')
data.head()
#total 400 values in our dataset
```

```python
#age , estimated sal : i/p , inde
#purchase : o/p
real_x = data.iloc[:, 2:4].values
real_y = data.iloc[:, 4].values
```

```python
#train_test(independent, depenet var)
train_x, test_x, train_y, test_y= train_test_split(real_x, real_y, test_size = 0.25, random_state = 0)
train_x
```

```python
#feature scaling
#since difference between salary and age is very large so we need to perform feature scalling
#operation, feature scaling divides our independent and dependent var 
#into range of +-2, so that our prediction and calculation becomes easy
#in feature scaling our data get distributed in the range of +-2
scaler = StandardScaler()
#scaling training data
train_x = scaler.fit_transform(train_x)
test_x = scaler.fit_transform(test_x)
train_x


```

```python
#creating model
classifier_LR = LogisticRegression(random_state = 0)
classifier_LR.fit(train_x, train_y)
```

```python
#validation
y_pred = classifier_LR.predict(test_x)
```

```python
y_pred
#o means no, 1 means yes
```

```python
#checking
test_y
#there are some prediction
```

```python
#confusion matrix helps us to find our wrong and right prediction done by our 
#modal so that we can avoid manual efforts
#y_true = actual result in our dataet
#y_pred = predicted value by our dataset, o/p
c_m = confusion_matrix(test_y, y_pred)
# we have total test data 25% that is 100 = 63+5+8+24
#digonal value are predicted correct = 63+24
#antidiagonal are total wrong prediction = 8+5
#accuracy = 87
#here we are using linear classifier so accuracy is a bit low
#why it is linear : we are calling logistic reg linear becasue
#it's plot is linear, and it uses linear data
#logistic reg need linear data
c_m
```

```python
#plotting logistic regression
#here our graph is diff because we have feature scaled to bring our data between +-2
from matplotlib.colors import ListedColormap
#we don't want our training data to get affected so we have stored them
#in two different copy
x_set, y_set = train_x, train_y
#to take the distribution of our data 
#we have splitted our data in +2 and -2
#X1 = min, X2 = max
#meshgrid helps in distribution of our data
#we are pointing points so we need matrix , start , end ,resolution
#koi point -2 pe directly aa skta toh -1 taki wo pura pura graph me

#visible ho
#step resolution 0.01

#to plot points here we will use contourf will help us
#to define data of x map and y map divide data

#ravel function finds out mid point
X1, X2 = np.meshgrid(np.arange(start = x_set[:,0].min() -1, stop = x_set[:, 0].max()+1, step = 0.01),
                    np.arange(start = x_set[:,1].min() -1, stop = x_set[:, 1].max()+1, step = 0.01))
plt.contourf(X1, X2, classifier_LR.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
                    
    
#logistic regression 
#will draw scaler plot with help of x and y variable
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
               c = ListedColormap(('red', 'green'))(i), label = j)

plt.title("Logistic Regression (Training set)")
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

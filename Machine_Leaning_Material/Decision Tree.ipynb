```python
#decision tree uses non linear data same as polynomial regression so using the same data set
import numpy as np
import pandas as pd
#building linear model
from sklearn.linear_model import LinearRegression
#for applying polynomial line reg
from sklearn.preprocessing import PolynomialFeatures
#starting to plot
import matplotlib.pyplot as plt
#for decesion tree
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv(r'/Users/divyashahi/OneDrive - Adobe/Machine_Leaning_Material/Position_Salaries.csv')
data.head(10)
```

```python
#doing prediction of this salary
real_x = data.iloc[:, 1:2].values # we just want first row
real_y = data.iloc[:, 2].values #dependent var
```

```python
#since our dataset is small we won't split here in training and testing dataset
#directly preparing model
reg = DecisionTreeRegressor(random_state=0)
#doing model train
reg.fit(real_x, real_y)
real_y
```

```python
#doing prediction
y_pred = reg.predict(real_x)
y_pred
```

```python
#doing prediction
y_pred = reg.predict(np.array([6.5]).reshape(-1,1))
y_pred
```

```python
#plotting the graph
#for accuracy taking step as 0.01
x_grid = np.arange(min(real_x), max(real_x), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
#doing scatter plot to rep real time data
plt.scatter(real_x, real_y, color = 'green')
plt.plot()
```

```python
#plotting the graph
#for accuracy taking step as 0.01
x_grid = np.arange(min(real_x), max(real_x), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
#doing scatter plot to rep real time data
plt.scatter(real_x, real_y, color = 'green')
plt.plot(x_grid, reg.predict(x_grid), color = 'blue')
plt.title("Decision tree regressor")
plt.xlabel("Pos level ")
plt.ylabel("Salary")
plt.show()
```
